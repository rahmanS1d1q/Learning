Soal Cerita: Perhitungan Gaji Karyawan (Nested If)

---

Sebuah perusahaan ingin menghitung **gaji bersih karyawan** berdasarkan beberapa aturan berikut.

Data Karyawan
Setiap karyawan memiliki data:
- `gaji_pokok`
- `lama_kerja` (dalam tahun)
- `status_karyawan` → `"tetap"` atau `"kontrak"`
- `menikah` → `True` atau `False`
- `jumlah_anak`

---

Aturan Perusahaan

Status Karyawan
- Jika **status karyawan adalah "tetap"**:
  - Karyawan **berhak atas tunjangan**
- Jika **status karyawan adalah "kontrak"**:
  - **Tidak mendapatkan tunjangan**
  - Langsung ke perhitungan gaji bersih

---

Lama Kerja (Hanya untuk Karyawan Tetap)
- Jika `lama_kerja` **≥ 5 tahun**:
  - Tunjangan masa kerja **10% dari gaji pokok**
- Jika `lama_kerja` **< 5 tahun**:
  - Tunjangan masa kerja **5% dari gaji pokok**

---

Status Pernikahan (Hanya untuk Karyawan Tetap)
- Jika `menikah == True`:
  - Mendapat tunjangan keluarga:
    - **Rp300.000** jika memiliki **anak**
    - **Rp200.000** jika **belum memiliki anak**
- Jika `menikah == False`:
  - Tidak mendapatkan tunjangan keluarga

---

Pajak
- Jika **gaji kotor ≥ Rp7.000.000** → pajak **10%**
- Jika **gaji kotor < Rp7.000.000** → pajak **5%**

---

Tugas
Buatlah program Python menggunakan **nested if** untuk:
1. Menghitung **gaji kotor**
2. Menghitung **pajak**
3. Menghitung **gaji bersih**
4. Menampilkan rincian gaji karyawan

---

Contoh Data

```python
gaji_pokok = 6_000_000
lama_kerja = 6
status_karyawan = "tetap"
menikah = True
jumlah_anak = 2

````python
# Data karyawan
gaji_pokok = 6_000_000
lama_kerja = 6                 # tahun
status_karyawan = "tetap"      # "tetap" / "kontrak"
menikah = True
jumlah_anak = 2

# Inisialisasi
tunjangan_masa_kerja = 0
tunjangan_keluarga = 0

# Status karyawan
if status_karyawan == "tetap":
    print("Status karyawan: Tetap")

    # Lama kerja
    if lama_kerja >= 5:
        kalkulasi = 0.10 * gaji_pokok
        tunjangan_masa_kerja = gaji_pokok + kalkulasi
    else:
        kalkulasi = 0.05 * gaji_pokok
        tunjangan_masa_kerja = gaji_pokok + kalkulasi

    # Status pernikahan
    if menikah:
        if jumlah_anak > 0:
            tunjangan_keluarga = 300_000
        else:
            tunjangan_keluarga = 200_000
    else:
        tunjangan_keluarga = 0

else:
    print("Status karyawan: Kontrak")
    tunjangan_masa_kerja = 0
    tunjangan_keluarga = 0

# Hitung gaji kotor
gaji_kotor = gaji_pokok + tunjangan_masa_kerja + tunjangan_keluarga

# Pajak
if gaji_kotor >= 7_000_000:
    pajak_persen = 0.10
else:
    pajak_persen = 0.05

pajak = gaji_kotor * pajak_persen

# Gaji bersih
gaji_bersih = gaji_kotor - pajak

# Output
print("Gaji pokok:", gaji_pokok)
print("Tunjangan masa kerja:", int(tunjangan_masa_kerja))
print("Tunjangan keluarga:", tunjangan_keluarga)
print("Gaji kotor:", int(gaji_kotor))
print(f"Pajak: {int(pajak_persen * 100)}%")
print("Gaji bersih:", int(gaji_bersih))
````

### `break`, `continue`, dan `pass`
---

`break` dan `continue` digunakan untuk **mengontrol alur perulangan** pada `for` dan `while`.

---


`break` digunakan untuk **menghentikan loop sepenuhnya**, meskipun kondisi loop masih memungkinkan untuk berjalan.

Struktur Dasar:
```python
for item in iterable:
    if kondisi:
        break
```

`continue` digunakan untuk melewati satu iterasi dan langsung melanjutkan ke iterasi berikutnya, tanpa menghentikan loop.

Struktur Dasar:
```python
for item in iterable:
    if kondisi:
        continue
```

`pass` digunakan sebagai placeholder atau pernyataan kosong.
Digunakan saat struktur loop sudah benar, tetapi belum ada aksi yang ingin dijalankan.

Struktur Dasar:
```python
for item in iterable:
    if kondisi:
        pass
```

Soal:
- Hentikan perulangan saat angka `0` ditemukan.

````python
data = [5, 3, 2, 0, 9]

# iterasi
for x in data:

    # kondisi
    if x == 0:
        break # stop
    print(x)
````

Soal:
- Lewati nilai yang kurang dari 60.

````python
scores = [80, 55, 90, 40, 70]

# iterasi
for s in scores:

    # kondisi
    if s < 60:
        continue # lanjutkan
    print(s)
````

Soal:
- Siapkan struktur perulangan untuk validasi yang belum dibuat.

````python
data = [1, 2, 3, 4]

for x in data:
    if x == 3:
        pass
    else:
        print(x)
````

## Function
---

`Function` digunakan untuk **mengelompokkan kode** agar bisa dipakai ulang dan membuat program lebih rapi.

---

`print` vs `return`

`print` digunakan untuk **menampilkan output ke layar**.

- Hanya untuk **melihat hasil**
- Nilainya **tidak bisa dipakai lagi**
- Biasanya untuk debugging atau tampilan

Contoh konsep:
> “Tampilkan hasilnya saja.”

---

`return` digunakan untuk **mengembalikan nilai dari function**.

- Nilainya **bisa disimpan**
- Bisa diproses lagi
- Menghentikan eksekusi function

Contoh konsep:
> “Kembalikan hasilnya supaya bisa dipakai.”

---

Perbedaan Inti

| `print` | `return` |
|------|--------|

````python
def maksimum(data, ambil_indeks=False):
    """
    Mencari nilai maksimum dari sebuah list.

    Parameters
    ----------
    data : list of int or float
        List yang berisi angka-angka.

    ambil_indeks : bool, optional
        Jika True, return tuple (nilai, indeks). Default False.

    Returns
    -------
    int or float or tuple
        Nilai maksimum dari list, atau tuple (nilai, indeks)
        jika ambil_indeks=True.
    """
    nilai_maks = max(data)

    if ambil_indeks:
        indeks = data.index(nilai_maks)
        return nilai_maks, indeks

    return nilai_maks

# Panggil function
data = [10, 25, 7, 30, 18]

# Ambil nilai maksimum saja
hasil = maksimum(data)
print(hasil)
````

### OOP 1
---

Apa itu OOP?

**Object-Oriented Programming (OOP)** adalah paradigma pemrograman yang mengorganisasi kode ke dalam **class** dan **object**, sehingga data dan perilaku digabung dalam satu struktur.

---

Komponen Dasar OOP

Class adalah definisi struktur object yang berisi:
- attribute (data)
- method (fungsi)

```python
class User:
    pass


````python
class User:
  pass

# instance
user1 = User()
````

Spesial Method:

| Special Method | Dipanggil Saat | Kegunaan |
|---------------|---------------|----------|
| `__init__` | Object dibuat | Inisialisasi attribute |
| `__str__` | `print()` / `str()` | Representasi string user-friendly |
| `__repr__` | `repr()` | Representasi resmi (debugging) |
| `__eq__` | `==` | Perbandingan object |
| `__lt__` | `<` | Perbandingan urutan |
| `__len__` | `len()` | Panjang object |
| `__getitem__` | Indexing | Akses elemen |
| `__iter__` | `for` loop | Membuat object iterable |
| `__next__` | Iterasi | Mengambil elemen berikutnya |
| `__call__` | `obj()` | Object dapat dipanggil seperti fungsi |
| `__enter__` | `with` | Masuk context manager |
| `__exit__` | Keluar `with` | Keluar & cleanup context |
| `__add__` | `+` | Operator penjumlahan |
| `__hash__` | `hash()` | Digunakan di `set` / `dict` key |

---
Reference:
- https://docs.python.org/3/reference/datamodel.html#special-method-names

````python
class User:
    def __init__(self, name, age):
        # Dipanggil saat object dibuat
        self.name = name  # instance attribute
        self.age = age   # instance attribute

    def __str__(self):
        # Dipanggil oleh print()
        return f"User(name={self.name}, age={self.age})"

    def __repr__(self):
        # Representasi resmi (debug)
        return f"User('{self.name}', {self.age})"

    def __eq__(self, other):
        # Dipanggil saat menggunakan ==
        return self.name == other.name and self.age == other.age

    def __len__(self):
        # Dipanggil oleh len()
        return self.age

    def __call__(self):
        # Object bisa dipanggil seperti fungsi
        print(f"Hello, {self.name}")
````

````python
# instance
u1 = User("Andi", 25)
u2 = User("Andi", 25)

print(u1)            # __str__
print(repr(u1))      # __repr__
print(u1 == u2)      # __eq__
print(len(u1))       # __len__
u1()                 # __call__
````

### OOP 2 - Child Class (Inheritance)
---

Inheritance adalah konsep OOP di mana **sebuah class (child)** mewarisi **attribute dan method** dari **class lain (parent)**.

---

Parent class adalah class dasar yang akan diwarisi.

```python
class User:
    def __init__(self, name):
        self.name = name

    def info(self):
        print(f"User name: {self.name}")
```
Child class adalah class yang mewarisi parent class menggunakan tanda kurung ()

```python
class Admin(User):
    def __init__(self, name, role):
        super().__init__(name)
        self.role = role
```
- Admin(User) → Admin adalah child, User adalah parent
- super().__init__(name) → memanggil constructor parent


````python
class User:
    def __init__(self, name):
        self.name = name

    def info(self):
        print(f"User name: {self.name}")

class Admin(User): # parent class dipanggil kesini
    def __init__(self, name, role):
        super().__init__(name)
        self.role = role
````

````python
admin = Admin(name = "Andi", role = "Admin")

admin.info()        # method dari parent
print(admin.role)   # attribute dari child
````

Soal:

Sebuah perusahaan ingin membuat sistem sederhana untuk menghitung gaji karyawan tetap.

Setiap karyawan memiliki data dasar:

- nama
- gaji pokok

Perusahaan hanya memiliki 1 jenis karyawan lanjutan, yaitu Karyawan Tetap.

Karyawan Tetap selain gaji pokok juga memiliki:

- jam lembur per bulan
- tarif lembur per jam
- bonus bulanan

Perusahaan juga menerapkan:
- pajak 10% dari total gaji kotor

````python
class Karyawan:
    def __init__(self, nama, gaji_pokok):
        """
        Inisialisasi data dasar karyawan.

        Parameters
        ----------
        nama : str
            Nama karyawan.

        gaji_pokok : int or float
            Gaji pokok per bulan.
        """
        self.nama = nama
        self.gaji_pokok = gaji_pokok

    def info(self):
        """
        Menampilkan informasi dasar karyawan.
        """
        print(f"Nama: {self.nama}")
        print(f"Gaji Pokok: {self.gaji_pokok}")


class KaryawanTetap(Karyawan):
    def __init__(self, nama, gaji_pokok, jam_lembur, tarif_lembur, bonus):
        """
        Inisialisasi data karyawan tetap.

        Parameters
        ----------
        nama : str
            Nama karyawan.

        gaji_pokok : int or float
            Gaji pokok per bulan.

        jam_lembur : int
            Total jam lembur per bulan.

        tarif_lembur : int or float
            Tarif lembur per jam.

        bonus : int or float
            Bonus bulanan.
        """
        super().__init__(nama, gaji_pokok)
        self.jam_lembur = jam_lembur
        self.tarif_lembur = tarif_lembur
        self.bonus = bonus

    def hitung_gaji_total(self):
        """
        Menghitung gaji bersih karyawan tetap.

        - Lembur hanya dihitung jika jam_lembur > 0
        - Pajak 10% dikenakan jika gaji kotor >= 10.000.000

        Returns
        -------
        dict
            Dictionary berisi rincian gaji:
            - nama
            - gaji_pokok
            - lembur
            - bonus
            - pajak
            - gaji_bersih
        """
        # hitung lembur
        if self.jam_lembur > 0:
            gaji_lembur = self.jam_lembur * self.tarif_lembur
        else:
            gaji_lembur = 0

        # gaji sebelum pajak
        gaji_kotor = self.gaji_pokok + gaji_lembur + self.bonus

        # aturan pajak
        if gaji_kotor >= 10_000_000:
            pajak = 0.1 * gaji_kotor
        else:
            pajak = 0

        gaji_bersih = gaji_kotor - pajak

        return {
            "nama": self.nama,
            "gaji_pokok": self.gaji_pokok,
            "lembur": gaji_lembur,
            "bonus": self.bonus,
            "pajak": pajak,
            "gaji_bersih": gaji_bersih
        }
````

````python
# membuat object KaryawanTetap
karyawan = KaryawanTetap(
    nama="Andi",
    gaji_pokok=8_000_000,
    jam_lembur=12,
    tarif_lembur=75_000,
    bonus=1_500_000
)

# panggil method info (dari parent class)
karyawan.info()

# hitung gaji total
hasil = karyawan.hitung_gaji_total()

# tampilkan hasil perhitungan
print("\nRincian Gaji:")
for key, value in hasil.items():
    print(f"{key}: {value}")
````
